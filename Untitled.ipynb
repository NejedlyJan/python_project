```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

```python
sales = pd.read_csv("videogames_sales.csv")
sales.head()
```

```python
sales.info()
```

```python
sales.describe()
```

```python
sales.isnull.sum()
```

```python
sales['Year'] = sales['Year_of_Release'].astype(int)
```

```python
plt.figure(figsize=(40,20))
sns.set(font_scale = 3)
sns.heatmap(sales.isnull(),yticklabels=False,cbar=False,cmap='magma')
```

```python
plt.figure(figsize=(40,50))
sns.set(font_scale = 1)
sns.jointplot(x="Year_of_Release", y="NA_Sales", data=sales)
```

```python
plt.figure(figsize=(40,20))
sns.set(font_scale = 3)
sns.countplot(x='Genre', data=sales, order = sales['Genre'].value_counts().index, palette="magma")
```

```python
plt.figure(figsize=(20,10))
sns.set(font_scale = 2)
sns.heatmap(sales.corr())
```

```python
plt.figure(figsize=(50,30))
sns.countplot(x='Developer',data=sales)
plt.xticks(rotation=90)
```
