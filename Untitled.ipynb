```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

```python
sales = pd.read_csv("videogames_sales.csv")
sales.head()
```

```python
from sklearn.model_selection import train_test_split
```

```python
X = sales.drop(['NA_Sales','EU_Sales', 'JP_Sales', 'Other_Sales', 'Global_Sales'],axis=1)
y = sales['Global_Sales']
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.30)
```

```python
sales = sales.dropna()
columns = list(sales)
columns.remove('NA_Sales')
columns
```

```python
X = sales.drop(['Name','EU_Sales', 'JP_Sales', 'NA_Sales','Other_Sales', 'Global_Sales', 'Rating', 'Developer', 'Publisher', 'Platform', 'Genre'], axis=1)
y = sales['NA_Sales']
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.30)
```

```python
from sklearn.linear_model import LinearRegression
```

```python
X
```

```python
lm = LinearRegression()
lm.fit(X_train,y_train)
```

```python
print(lm.intercept_)
```

```python
coeff_df = pd.DataFrame(lm.coef_,X.columns,columns=['Coefficient'])
coeff_df
```

```python
predictions = lm.predict(X_test)
```

```python
from sklearn import metrics
print('MAE:', metrics.mean_absolute_error(y_test, predictions))
print('MSE:', metrics.mean_squared_error(y_test, predictions))
print('RMSE:', np.sqrt(metrics.mean_squared_error(y_test, predictions)))
```
