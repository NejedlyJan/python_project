```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

```python
sales = pd.read_csv("videogames_sales.csv")
sales.head()
```

```python
sales.info()
```

```python
sales.describe()
```

```python
sns.heatmap(sales.isnull(),yticklabels=False,cbar=False,cmap='plasma')
```

```python
sns.jointplot(x="Year_of_Release", y="NA_Sales", data=sales)
```

```python
sns.distplot(sales['Genre'],bins=2,kde=False,color='red')
```

```python
sns.heatmap(sales.corr())
```

```python
plt.figure(figsize=(50,30))
sns.countplot(x='Developer',data=sales,order = sales['Developer'].value_counts().index)
plt.xticks(rotation=90)
```

```python
clean_sales = sales.dropna
sns.jointplot(x='NA_Sales',y='Critic_score',data=clean_sales,color='red',kind='kde');
```
